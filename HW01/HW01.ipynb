```python
from lecture02_demos import demo01, demo02, demo04
import matplotlib.pyplot as plt
import numpy as np
```

1-(1)   Lax-Wendroff scheme
        Demonstrate (numerically) that it is second-order accurate

```python
n = np.linspace(100,400,7)
err_LaxWendroff = []
for i in range(0,7):
    err_LaxWendroff.append(demo02.Lax_Wendroff(N=int(n[i])))

err_LaxWendroff = np.array(err_LaxWendroff)

a, b = np.polyfit(np.log10(n), np.log10(err_LaxWendroff), 1)
plt.plot(np.log10(n), np.log10(err_LaxWendroff), ls='-', c='k', marker='.', label='Lax Wendroff')
plt.plot(np.log10(n), a*np.log10(n)+b-0.1, ls='--', c='g', label=str(format(a,'.2f'))+'x + constant')
plt.xlabel('N [log]')
plt.ylabel('ERROR [log]')
plt.legend(frameon=False)
plt.show()
```

1-(2)   Lax-Wendroff scheme
        Compare (numerically) the order of accuracy with the upwind scheme

```python
err_Upwind = []
for i in range(0,7):
    err_Upwind.append(demo01.Upwind(N=int(n[i])))

err_Upwind = np.array(err_Upwind)
a, b = np.polyfit(np.log10(n), np.log10(err_LaxWendroff), 1)
c, d = np.polyfit(np.log10(n), np.log10(err_Upwind), 1)
plt.plot(np.log10(n), np.log10(err_LaxWendroff), ls='-', c='k', marker='.', label='Lax Wendroff')
plt.plot(np.log10(n), np.log10(err_Upwind), ls='-', c='b', marker='.', label='Upwind')
plt.plot(np.log10(n), a*np.log10(n)+b-0.1, ls='--', c='g', label=str(format(a,'.2f'))+'x + constant')
plt.plot(np.log10(n), c*np.log10(n)+d-0.1, ls='--', c='r', label=str(format(c,'.2f'))+'x + constant')
plt.xlabel('N [log]')
plt.ylabel('ERROR [log]')
plt.legend(frameon=False)
plt.show()
```
